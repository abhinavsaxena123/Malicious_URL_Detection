### Model Building

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score,classification_report,confusion_matrix, ConfusionMatrixDisplay

def evaluate_model(y_true, y_pred, model_name="Model", average='macro'):
    
    # Calculating metrics
    accuracy = round(accuracy_score(y_true, y_pred) * 100, 2)
    f1 = round(f1_score(y_true, y_pred, average=average,zero_division=0) * 100, 2)
    recall = round(recall_score(y_true, y_pred, average=average,zero_division=0) * 100, 2)
    precision = round(precision_score(y_true, y_pred, average=average,zero_division=0) * 100, 2)

    # Printing results
    print(f"{model_name} Evaluation Metrics:")
    print(f"Accuracy: {accuracy}%")
    print(f"F1 Score: {f1}%")
    print(f"Recall: {recall}%")
    print(f"Precision: {precision}%")
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, zero_division=0))

    return {
        "accuracy": accuracy,
        "f1_score": f1,
        "recall": recall,
        "precision": precision,
    }
```

```python
# Function to store metrics score for each model...
model_score = {}
```

```python
def plot_confusion_matrix(cm,title='Confusion Matrix',cmap=plt.cm.Reds,display_metrics=True):
    plt.figure(figsize=(7, 5))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    
    #0 -> Benign | 1 -> Defacement | 2 -> Malware | 3 -> Phishing
    classes = ['Benign', 'Defacement', 'Malware', 'Phishing'] 
    plt.xticks(ticks=np.arange(len(classes)), labels=classes)
    plt.yticks(ticks=np.arange(len(classes)), labels=classes)

     # Add text annotations to each cell
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, f'{cm[i, j]}',
                     ha='center', va='center',
                     color='white' if cm[i, j] > cm.max() / 2. else 'black')
            
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.grid(False)
```

### 1. Multinomial Naive Baiyes

* The Multinomial Naive Bayes model is designed to work with discrete, non-negative feature values. Since the tld_length column contains negative value, it would not be appropriate to feed it into a Multinomial Naive Bayes model.

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
X_df = X_train.drop(['tld_length'], axis = 1)
X_test_df = X_test.drop(['tld_length'], axis = 1)
```

```python
mnb = MultinomialNB()
mnb.fit(X_df, Y_train)
```

```python
# Make predictions
y_pred_mnb= mnb.predict(X_test_df)

metrics = evaluate_model(Y_test, y_pred_mnb, model_name="MNB")
model_score["MNB"] = list(metrics.values())
```

```python
mnb_cm = confusion_matrix(Y_test, y_pred_mnb)
plot_confusion_matrix(mnb_cm)
```

### 2. Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(max_depth=6)
dtc.fit(X_train, Y_train)
```

```python
# Make predictions
y_pred_dtc = dtc.predict(X_test)

metrics = evaluate_model(Y_test, y_pred_dtc, model_name="Decision Tree")
model_score["Decision Tree"] = list(metrics.values())
```

### Using Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'criterion': ['gini', 'entropy'],       # Criteria to split
    'max_depth': [None, 5, 6, 8, 10], # Maximum depth of the tree
    'min_samples_split': [2, 5, 10], # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4],    # Minimum number of samples required to be at a leaf node
    'max_features': [None, 'sqrt', 'log2'] # Number of features to consider when looking for the best split
}

dtc = DecisionTreeClassifier()

grid_search = GridSearchCV(estimator=dtc, param_grid=param_grid, 
                           cv=5, n_jobs=-1, verbose=2, scoring='accuracy')

grid_search.fit(X_train, Y_train)
```

```python
# Get the best estimator
best_dtc = grid_search.best_estimator_
print(best_dtc)

# Make predictions using the best estimator
y_pred_dtc = best_dtc.predict(X_test)
```

```python
metrics = evaluate_model(Y_test, y_pred_dtc, 
                         model_name="Decision Tree Classifier with GridSearchCV")

model_score["Decision Tree Classifier with GridSearchCV"] = list(metrics.values())
```

```python
dtc_cm = confusion_matrix(Y_test, y_pred_dtc)
plot_confusion_matrix(dtc_cm)
```

1. An accuracy of 91% is quite high, indicating that the model performs well on the entire dataset.
2. For Class 3 also recall and f1-score increased.

### ANN

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers, Sequential
from tensorflow.keras import layers, models, optimizers, losses
```

```python
model_ANN = Sequential([
    layers.InputLayer(input_shape=(X_train.shape[1],)), 
    
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),  # Dropout to prevent overfitting
    
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.3),
    
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),
    
    layers.Dense(4, activation='softmax')  # 4 output classes with softmax
])
```

```python
model_ANN.compile(
    optimizer=optimizers.Adam(),
    loss=losses.SparseCategoricalCrossentropy(),
     metrics=['accuracy']
)
model_ANN.summary()
```

```python
history = model_ANN.fit(X_train, Y_train, epochs=10, batch_size=32, validation_split=0.2, verbose=1)
```

```python
# Retrieving training and validation metrics
history_dict = history.history
epochs = range(1, len(history_dict['loss']) + 1)
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs, history_dict['loss'], 'bo', label='Training loss')
plt.plot(epochs, history_dict['val_loss'], 'b', label='Validation loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Plot training & validation accuracy values
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 2)
plt.plot(epochs, history_dict['accuracy'], 'bo', label='Training accuracy')
plt.plot(epochs, history_dict['val_accuracy'], 'b', label='Validation accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
y_pred_prob_ann = model_ANN.predict(X_test)
predicted_classes = y_pred_prob_ann.argmax(axis=1)
```

```python
metrics = evaluate_model(Y_test, predicted_classes, model_name="ANN")
model_score["ANN"] = list(metrics.values())
```

```python
ann_cm = confusion_matrix(Y_test, predicted_classes)
plot_confusion_matrix(ann_cm)
```
